```python
#  use dict for this , OOPS concept should be implemented
#    here
#    input roll,name,avg. roll is the Key and value
#    is name,avg and result - say as list
#    compute the result based on avg as following :
#    1-39 - Fail
#    40-49 - Pass
#    50-59 - Second class
#    60-74- First class
#    75-89 - distinction
#    90 and above - High Distinction
#    have methods like 
#    input_data()
#    process_avg()
#    display_all_details()
#    etc
#    Find out how many candidates secures more than class
#    avg.
#    display the student details who has highest and second highest
#    avg
class Student:
    def __init__(self):
        print("Student Details")
    def input_data(self):
        self.student_dict={}
        n=int(input("Enter a how many Students :"))
        for i in range(0,n):
            roll=int(input("Enter a Roll Number : "))
            self.name=input("Enter a Student Name : ")
            avg=int(input("Enter a Student Average : "))
            self.student_dict.update({roll:[self.name,avg]})
        print("Student Datiles are : ",self.student_dict)
    def process_avg(self):
        self.names=[]
        self.avg=[]
        s=c=0
        for k,v in self.student_dict.items():
            self.names.append(v[0])
            self.avg.append(v[1])
        print("Students Names : ",self.names)
        print("Student Averages : ",self.avg)
        for j in self.avg:
            s=s+j
            c=c+1
            if 1<=j<=39:
                print("Your are Fail")
            elif 40<=j<=49:
                print("Your are Pass")
            elif 50<=j<=59:
                print("Your are Second Class")
            elif 60<=j<=74:
                print("Your are First Class ")
            elif 75<=j<=89:
                print("Your are Distinction ")
            elif j>=90:
                print("Your are High Distinction ")
        self.class_avg=s//c
    def above_class_avg(self):
        self.c1=0
        for h in self.avg:
            if h>self.class_avg:
                self.c1=self.c1+1
        
        self.avg.sort()
    def display_all_details(self):
        print("Class Average :",self.class_avg )
        print("Max Avg : ",self.avg[-1])
        print("Second Highest Avg : ",self.avg[-2])
        print("How Many Students are there above the class Average : ",self.c1)
s=Student()
s.input_data()
s.process_avg()
s.above_class_avg()
s.display_all_details()
```

```python
# input a number and display one of the following
#    it is a SINGLE digit number ::: if it is 1 to 9
#    it is DOUBLE digit number ::: if it is 10 to 99
#    and so on...
#    do it till SIX digit number
#    if number is above SIX digits then display this number is LARGE number
#    NOTE
#    if the user inputs a NEGATIVE number then - donot do the above processing
#    instead display :: cannot process NEGATIVE numbers

class Number:
    def input_number(self,num):
        self.num=num
        print("Number is : ",self.num)
    def calc(self):
        if 1<=self.num<=9:
            print("This is Single Digit Number")
        elif 10<=self.num<=99:
            print("This is Double Digit Number")
        elif 100<=self.num<=999:
            print("This Three Digit Number")
        elif 1000<=self.num<9999:
            print("This is Foure Digit Number")
        elif 10000<=self.num<99999:
            print("This is Five Digit Number")
        elif 100000<=self.num<999999:
            print("This is Six Digit Number")
        elif self.num>=1000000:
            print("It is a Large Number")
        else:
            if self.num<0:
                print("cannot process NEGATIVE numbers")
n1=int(input("Enter a Number : "))
n=Number()
n.input_number(n1)
n.calc()
                      

        
```

```python
#  input any 5 numbers and store in set
#    find the highest number
#    find the lowest number
#    find the avg of all the numbers
#    display the set in DESC order of numbers
#    convert the set to the list - call it list1
#    given list2 = [89,23,67,22,100,105,15]
#    process the list2 and add it into the list1
#    in the last 5 elements find the highest
#    in the first 6 elements find the lowest
#    find the avg of all the list1 elements
#    how many numbers are above the avg ?
class Number:
    def __init__(self):
        self.set1=set()
        n1=int(input("Enter a how many numbers : "))
        for i in range(1,n1+1):
            num=int(input("Enter a Number : "))
            if num=='Stop' or num=='STOP':
                break
            else:
                self.set1.add(num)
        print("set of the Numbers : ",self.set1)
    def highest_number(self):
        h=0
        for i in self.set1:
            if i>h:
                h=i
        print("Highest Value : ",h)
    def lowest_value(self):
        l=999999
        for j in self.set1:
            if j<l:
                l=j
        print("Lowest Value : ",l)
    def desc_order(self):
        print("After sorted the Set : ",sorted(self.set1,reverse=True))
    def convert_to_list(self):
        self.list1=list(self.set1)
        self.list2=[89,23,67,22,100,105,15,78,56,20]
        print("List1 Values : ",self.list1)
        print("List2 Values :  ",self.list2)
    def excended_list(self):
        self.list1.extend(self.list2)
        print(self.list1)
    def first_6_number(self):
        h1=0
        for h in self.list1[0:5]:
            if h>h1:
                h1=h
        print("First 6 Numbers in highest value : ",h1)
    def last_5_number(self):
        l1=999999
        for x in self.list1[-5:-1]:
            if x<l1:
                l1=x
        print("Last 5 Numbers in Lowest Value : ",l1)
    def avg_list(self):
        ln=len(self.list1)
        s=sum(self.list1)
        self.avg=s//ln
        print("Average of list : ",self.avg)
    def above_avg_elements(self):
        c1=0
        for y in self.list1:
            if y>self.avg:
                c1=c1+1
        print("how many numbers are there above the avg : ",c1)
n=Number()
n.highest_number()
n.lowest_value()
n.desc_order()
n.convert_to_list()
n.excended_list()
n.first_6_number()
n.last_5_number()
n.avg_list()
n.above_avg_elements()
```
